# pandas : parallel coordinate


```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

```python
matplotlib.rcParams['figure.figsize'] = (10, 10)
pd.plotting.parallel_coordinates(iris_df, 'target')
```

```python
iris_df['target'].value_counts()
```

# pandas : scatter matrix

```python
pd.plotting.scatter_matrix(iris_df)
```

#Color

```python
matplotlib.rcParams['figure.figsize'] = (10, 10)

plt.imshow(iris_df.iloc[:,:],interpolation='nearest',cmap='rainbow')
```

```python
matplotlib.rcParams['figure.figsize'] = (10, 10)
pd.plotting.parallel_coordinates(iris_df, 'target',color=('c','m','y')); # c m y k r g b w
```

```python
matplotlib.rcParams['figure.figsize'] = (10, 10)
pd.plotting.parallel_coordinates(iris_df, 'target',color=[(1,0,0),(0.5,0.5,0.5),(0,1,1)]); # c m y k r g b w
```

```python
matplotlib.rcParams['figure.figsize'] = (10, 10)
pd.plotting.parallel_coordinates(iris_df, 'target',color=['#375878','#783760','#D4C600']); # c m y k r g b w
```
